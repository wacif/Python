# Plotting in Python

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
titanic_data = sns.load_dataset('titanic')
titanic_data.head(5)
```

```python
titanic_data.info()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x = 'survived', hue = 'sex', data=titanic_data)
```

```python
# scatter plot
plt.scatter(x='age', y='fare', data=titanic_data)
```

```python
# scatter plot
sns.scatterplot(x='age', y='fare', data=titanic_data, hue='pclass')
```

```python
sns.barplot(x='sex', y='fare', data=titanic_data, hue='pclass')
```

```python
titanic_data.columns
```

```python
sns.pairplot(titanic_data)
```

```python
titanic_data.hist(figsize=(10, 8), grid=False)
```

```python
sns.boxplot(data=titanic_data, x='sex', y='fare', hue='pclass')
```

```python
sns.boxenplot(data=titanic_data, x='sex', y='fare', hue='pclass')
#how to change the title, x axis and y axis
```

```python
g = sns.FacetGrid(titanic_data, hue="survived", col='pclass', margin_titles=True,
                  palette={1: 'seagreen', 0: 'gray'})
g = g.map(plt.scatter, "fare", 'age', edgecolor= 'w').add_legend();
```

```python
sns.stripplot(data=titanic_data, x='sex', y='fare', hue='pclass')
```

```python
# heatmap

corr = titanic_data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```
