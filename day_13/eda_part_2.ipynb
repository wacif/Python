```python
# visualize and explore
sns.countplot(data=df, x='survived')
```

```python
sns.barplot(data=df, x='survived', y='fare', hue='who')
```

```python
sns.countplot(x='pclass', data=df)
```

```python
# data  di8tribution
sns.histplot(data=df, x='age', bins=32)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0)
```

```python
sns.boxplot(data=df, y='age', x='survived', hue='who')
```

```python
# scatterplot
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim(0,250)
```

```python
sns.pairplot(df, hue='survived')
```

```python
# stacked barchat/plot
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y='fare')
```

```python
sns.swarmplot(data=df, x='survived', y='age', hue='who')
```

```python
# Filter data for fares less than 300
filtered_df = df[df['fare'] < 300]
```

```python
# releation of two numeric variables
sns.lmplot(data=filtered_df, x='fare', y='age', hue='pclass')
```

```python
# how to add a column and make a age group
sns.histplot(data=df, x='age')
```

```python
# grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'],[0, 10, 18, 30, 50, 80], 
                          labels=['kids', 'child', 'youth', 'middle age', 'old'])
```

```python
df.head()
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```
