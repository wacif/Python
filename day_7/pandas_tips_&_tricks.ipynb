#   03- How to rename columns

```python
df = pd.DataFrame({'fruite_name': ["mango","banana","apple"], 'seaseon':["summer","winter","autumn"], 'ratings':[3.0,2.1,4]})
df
```

```python
df.rename(columns={'fruite_name': 'f_name', 'seaseon': 'season'}, inplace=True)
df
```

```python
# renamae columns 
df.columns = ['fruite_name','season','rating']
df
```

```python
# to replace any character or string
df.columns = df.columns.str.replace('fruite_name','f_name')
df
```

```python
df.columns = df.columns.str.replace('a',' ')
df
```

```python
# adding prefix to columns
df.columns=df.columns.str.replace(' ', 'a')
# df = df.add_prefix('-')
df.columns = ['f_name', 'seasons', 'ratings']
df
```

```python
count = dict()
names = ['a', 'b', 'c', 'a', 'a','g', 'b']

for name in names:
    if name not in count:
        count[name] = 1
    else:
        count[name] = count[name] + 1

print(count)
```

```python
count = dict()
names = ['a', 'b', 'c', 'a', 'a','g', 'b']

for name in names:
    count[name] = count.get(name, 0) + 1

print(count)
```
